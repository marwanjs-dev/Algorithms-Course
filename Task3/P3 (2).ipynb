## a) Algorithm Implementation

```python
# write your implementation here
def swap(list, pos1, pos2): 
    list[pos1], list[pos2] = list[pos2], list[pos1] 
    return list

arr = [1,2,3,4,5,6]
n = 6
s = int((n-1)/2)
e = int(n/2)
while s>=0 and e<= n-1:
        arr = swap(arr, s,e)
        s = s -1
        e= e + 1

print(arr)
```

## b) Growth Curve

Empirically, using *time measurements*, test your algorithm and **plot a graph** that shows its growth rate when variating the *size*.<br>

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import time
import random
```

```python
lst = []
start = 0
end = 1000
for i in range(50):
    rlist = random.sample(range(start, end), i+1)
    lst.append(rlist)
```

```python
times = []
for n in range(50):
    start_time = time.time()
    swap(lst[n], 0, len(lst[n])-1)
    execution_time_ns = (time.time() - start_time)*(10**9)
    times.append(execution_time_ns)
```

```python
# the code to plot the curve
f,ax = plt.subplots(figsize=(15, 10))
ax.plot(range(50),times)
ax.set_xlabel('$n$')
ax.set_ylabel('Time: nano seconds')
ax.legend(loc=0);
```

from this graph it is obvious that time is seemingly stable so the size of the list doesn't change the time much

# Induction Case:
We want to show that $$R(L) \implies R(L+1)$$ and L is the length of the list
* Therefore, assuming $R(L)$ is True:<br>
$$R(L) \implies \texttt{R}\rvert_L = R(A[0],A[1],\cdots,A[L-1])$$
For the next iteration $\texttt{new value} = A[L]$; then, we have a condition as follows
\begin{align*}
R(L) &\implies\\
\texttt{R}\rvert_{L+1}

&= R(\underbrace{(A[0],A[1],\cdots,A[L-1])}_{same values of R(L)},\underbrace{A[L]}_{new value})\\
&= R(A[0],A[1],\cdots,A[L])
\end{align*}
Therefore, after the end of this iteration the variable
$$\underbrace{\texttt{R(L+1)} = R(A[0],A[1],\cdots,A[L])}_{R(L+1)}$$
$$\boxed{R(L) \implies R(L+1)}$$
