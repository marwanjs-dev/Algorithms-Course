## Problem Statement

Considering the search problem, we have a list of $n$ integers $A = \langle v_1, v_2, \cdots v_n \rangle$. We want to design an algorithm to check whether an item $v$ exists or not such that it should return either the **index**, $i$, if it was found or $-1$ otherwise, when *not found*.

## a) Algorithm Implementation

```python
# write your implementation here
def search(A, target):
    for i in range(len(A)):
        if A[i] == target: return i
    return -1
```

## b) Growth Curve

Empirically, using *time measurements*, test your algorithm and **plot a graph** that shows its growth rate when variating the *size*.<br>

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import time
import random
```

```python
lst = []
randomlst = []
start = 0
end = 1000
for i in range(50):
    rlist = random.sample(range(start, end), i+1)
    lst.append(rlist)
    #random_num = random.choice(rlist)
    random_num = rlist[-1]
    randomlst.append(random_num)
```

```python
times = []
for n in range(50):
    start_time = time.time()
    search(lst[n], randomlst[n])
    # Total time = end - start
    execution_time_ns = (time.time() - start_time)*(10**9)
    times.append(execution_time_ns)
```

```python
# the code to plot the curve
f,ax = plt.subplots(figsize=(15, 10))
ax.plot(range(50),times)
ax.set_xlabel('$n$')
ax.set_ylabel('Time: nano seconds')
ax.legend(loc=0);
```
