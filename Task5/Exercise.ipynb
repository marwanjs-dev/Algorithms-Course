```python
# write your implementation here
from collections import deque

def bfsPath(graph,src,tgt):
    """
    it returns the first path found from <s> to <e>
    ----------------------------------------------------------------------------
    Inputs:
    - g: the graph representation
    - s: the start node
    - e: the goal node
    ----------------------------------------------------------------------------
    Output:
    - path: list contains ordered cities according to the path of  BFS algorithm
    ----------------------------------------------------------------------------
    """
    parents = {src: None}
    queue = deque([src])
    while queue:
        node = queue.popleft()
        for neighbor in graph[node]:
            if neighbor not in parents:
                parents[neighbor] = node
                queue.append(neighbor)
                if node == tgt:
                    break
    path = [tgt]
    while parents[tgt] is not None:
        path.insert(0, parents[tgt])
        tgt = parents[tgt]

    return path
```

```python
from copy import copy
def branch_and_bound(net, s, t):
    """
    The function returns the shortest path from <s> to <e> using the  branch and
    bound algorithm without any search pruning techniques.    
    ----------------------------------------------------------------------------
    Inputs:
    - g: the graph representation
    - s: the start node
    - e: the goal node
    ----------------------------------------------------------------------------
    Output:
    - shortest_path: list contains ordered cities according to the shortest path
    ----------------------------------------------------------------------------
    """
    labels={}
    order={}
    for i in net.keys():
        if i == s: labels[i] = 0 
        else: labels[i] = float("inf") 
    drop1 = copy(labels)
    while len(drop1) > 0:
        minNode = min(drop1, key = drop1.get) 
        for i in net[minNode]:
            if labels[i] > (labels[minNode] + net[minNode][i]):
                labels[i] = labels[minNode] + net[minNode][i]
                drop1[i] = labels[minNode] + net[minNode][i]
                order[i] = minNode
        del drop1[minNode]
    temp = copy(t)
    rpath = []
    path = []
    while 1:
        rpath.append(temp)
        if temp in order: temp = order[temp]   
        else: return "There is no path from " + str(s) + " to " + str(t) + "."
        if temp == s:
            rpath.append(temp)
            break
    for j in range(len(rpath)-1,-1,-1):
        path.append(rpath[j])
    return "The shortest path from " + s + " to " + t + " is " + str(path) + ". Minimum distance is " + str(labels[t]) + "."

```

```python
# Lets try it out
graph2 = romania_map2()
path = branch_and_bound(graph2,'A','U')
print(path)
```

### A Star

Now, modify your implementation to considering the heuristic, implementing the A star (A*) algorithm.

```python
def A_star(graph, heuristic, start, goal):
    """
    The function returns the shortest path from  <s>  to  <e> using the A* Algo-
    rithm with the use of a visited list along with the heuristic values.
    ----------------------------------------------------------------------------
    Inputs:
    - g: the graph representation
    - s: the start node
    - e: the goal node
    - h: some strcture holding the heuristic values to the goal from each node
    ----------------------------------------------------------------------------
    Output:
    - shortest_path: list contains ordered cities according to the shortest path
    ----------------------------------------------------------------------------
    """

    distances = dict(zip(graph.keys(), [float("inf")] * len(graph)))

    distances[start] = 0

    priorities = dict(zip(graph.keys(), [float("inf")] * len(graph)))

    priorities[start] = heuristic[start][goal]

    visited = dict(zip(graph.keys(), [False] * len(graph)))
    nodes = []
    nodes.append(start)
    while True:
        lowest_priority = float("inf")
        lowest_priority_index = -1
        for i in priorities.keys():
            if priorities[i] < lowest_priority and not visited[i]:
                lowest_priority = priorities[i]
                lowest_priority_index = i
        if lowest_priority_index == goal:
            nodes.append(goal)
            return distances[lowest_priority_index], nodes
        for i in graph[lowest_priority_index].keys():
            if graph[lowest_priority_index][i] != 0 and not visited[i]:
                if distances[lowest_priority_index] + graph[lowest_priority_index][i] < distances[i]:
                    if lowest_priority_index not in nodes and lowest_priority_index in graph[nodes[-1]].keys():
                        nodes.append(lowest_priority_index)
                    distances[i] = distances[lowest_priority_index] + graph[lowest_priority_index][i]
                    priorities[i] = distances[i] + heuristic[i][goal]
                visited[lowest_priority_index] = True
```
